```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("airquality.csv")

# 1. Mostrem les 10 primeres línies
print(df.head(10))

```

```python
print(f"Longitud del dataset: {len(df)}")
```

```python
print(df.info())
```

```python
max_temp = df['Temp'].max()
print(f"La temperatura màxima és: {max_temp}")
```

```python
# 5. Mes i dia amb la temperatura màxima
max_temp_row = df[df['Temp'] == max_temp]
print(f"Temperatura màxima registrada el dia {max_temp_row['Day'].values[0]} del mes {max_temp_row['Month'].values[0]}")
```

```python
# 6. Gràfic de punts: relació entre el vent i el Solar.R
plt.figure(figsize=(8, 6))
sns.scatterplot(x=df["Wind"], y=df["Solar.R"])
plt.xlabel("Velocitat del vent")
plt.ylabel("Radiació solar")
plt.title("Relació entre el vent i la radiació solar")
plt.show()
```

```python
# 7. Gràfic de barres: temperatura mitjana per mes
df.groupby("Month")["Temp"].mean().plot(kind="bar", color="skyblue", figsize=(8, 6))
plt.xlabel("Mes")
plt.ylabel("Temperatura mitjana")
plt.title("Temperatura mitjana per mes")
plt.show()
```

```python
# 8.Gràfic de punts: relació entre Ozó i Temperatura
plt.figure(figsize=(8, 6))
sns.scatterplot(x=df["Ozone"], y=df["Temp"])
plt.xlabel("Concentració d'Ozó")
plt.ylabel("Temperatura")
plt.title("Relació entre Ozó i Temperatura")
plt.show()
```
